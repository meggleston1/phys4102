Part 1(c)

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.loadtxt('sunspots.txt')                           # Loads in the data from the given file as an array

r = 5                                                    # Defines the needed variable 'r'
MovingAverage = np.zeros(len(a))
for index in range(0,len(a[:,1])):
    if (index-r) < 0:
        q = np.sum(a[index:index+r,1])
        MovingAverage[index] = q/(2*r)
    elif (index+r) > len(a):
        q = np.sum(a[index-r:index,1])
        MovingAverage[index] = q/(2*r)
    else:
        q = np.sum(a[index-r:index+r,1])
        MovingAverage[index] = q/(2*r)

plt.plot(MovingAverage,'r')
plt.scatter(a[:,0] , a[:,1] , 0.5)
```
