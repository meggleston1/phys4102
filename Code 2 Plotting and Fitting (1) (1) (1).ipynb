Part 1(c)

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.loadtxt('sunspots.txt')                        # Loads in the data from the given file as an array

r = 5                                                 # Defines the needed variable 'r'
MovingAverage = np.zeros(len(a))                      # Sets up an array of zeros with dimensions as long as array 'a'

for index in range(0,len(a[:,1])):                    # 'for' loop that sets up the running calculation
    
    if (index-r) < 0:                                   # Gives conditions for when the left index point is negative
        Diff_1 = abs(index-r)                             # Establsihes how far negative the left index point is
        Diff_2 = abs((Diff_1)-r)                          # Establishes how many data points left of the index are needed
        q = np.sum(a[index-Diff_2:index+r,1])             # Sets up 'q' as function that adds up the needed section of array 'a'
        MovingAverage[index] = q/(2*r)                    # Places proper value into the 'MovingAverage' array
        
    elif (index+r) > len(a):                            # Gives conditions for when the right index point is > len(a)
        Diff_3 = ((index+r)-(len(a)))                     # Establsihes how far above len(a) the right index point is
        Diff_4 = abs((Diff_3)-r)                          # Establishes how many data points right of the index are needed
        q = np.sum(a[index-r:index+Diff_4,1])             # Sets up 'q' as function that adds up the needed section of array 'a'
        MovingAverage[index] = q/(2*r)                    # Places proper value into the 'MovingAverage' array
        
    else:                                               # Instructions for all other cases
        q = np.sum(a[index-r:index+r,1])                  # Sets up 'q' as function that adds up the needed section of array 'a'
        MovingAverage[index] = q/(2*r)                    # Places proper value into the 'MovingAverage' array
        
plt.xlabel('Months Since January 1749')                  # Labels the axis
plt.ylabel('Number of Sunspots')
plt.title('Sunspot Data')                                # Gives the plot a title
plt.plot(MovingAverage,'r',label='Running Average')      # Plots the Running Average and establishes a label and color
plt.scatter(a[:,0],a[:,1],0.5,label='Sunspot Count')     # Displays the scatterplot and establishes a label
plt.legend()                                             # Adds a Legend based on the established labels
plt.show()                                               # Skips the output data lines
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0,np.pi,0.001)                       # Establsihes a value for time ('t')
freq_1 = 100                                       # First frequency
freq_2 = 105                                       # Second frequency
a = np.sin(freq_1*t)                               # Sine wave equation
b = np.sin(freq_2*t)

plt.figure(figsize=(22,5))                         # Establsihes a figure size that better shows the 2 sine waves      
plt.xlabel('Time (s)')                             # Labels the axis
plt.ylabel('Amplitude')
plt.title('Sine Waves')                            # Gives the plot a title
plt.plot(a,label='100 Hz')                         # Graphs the data and establishes labels
plt.plot(b,label='105 Hz')
plt.legend()                                       # Sets up a legend

plt.figure(figsize=(22,5))                         # Establsihes a figure size that better shows the resulting sine wave   
plt.xlabel('Time (s)')                             # Labels the axis
plt.ylabel('Amplitude')
plt.title('Interference')                          # Gives the plot a title
plt.plot(a+b)                                      # Graphs the data
plt.show()                                         # Skips the output data lines
```

Part 2(b)

```python
import numpy as np
import matplotlib.pyplot as plt

WLength = 5.0                                                 # Sets up a Wavelength variable
Amp = 1.0                                                     # Sets up an Amplitude variable
k = (2*np.pi)/WLength                                         # Equation for 'k' (Wavevector)

dC_C = 20.0                                                   # Distance (cm) between the pebble drops
SqSideLen = 100                                               # Distance (cm) of the edge of the graphed area
Div = 500                                                     # Number of divisions on each side of the graphed area
Den = SqSideLen/Div                                           # Distance between divisions on each side of the graphed area

x_coord_1 = SqSideLen/2 + dC_C/2                              # Origin of circle 1
y_coord_1 = SqSideLen/2
x_coord_2 = SqSideLen/2 - dC_C/2                              # Origin of circle 2
y_coord_2 = SqSideLen/2

q = np.zeros((Div,Div))                                       # Sets up an empty (0) array to fill with values

for a in range(Div):                                          # Establishes a loop to run through an array of values
    y = (Den)*a                                               # Assigs the variable 'y' to the results from the previous loop
    for b in range(Div):                                      # Establishes a loop to run through an array of values
        x = Den*b
        r_1 = np.sqrt((x-(x_coord_1))**2+(y-(y_coord_1))**2)  # Given equation for 'r'
        r_2 = np.sqrt((x-(x_coord_2))**2+(y-(y_coord_2))**2)
        q[a,b] = Amp*np.sin(k*r_1) + Amp*np.sin(k*r_2)        # Fills in array 'q' with the results
        
plt.imshow(q,cmap='gray')                                     # Displays the image in grayscale
plt.show()                                                    # Skips the output data lines
```

Part 5(d/e)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as spy

q = np.loadtxt('FreqVoltData')
Freq = q[:,0]                                          # Reads the data from array 'q' column zero
Volt = q[:,1]                                          # Reads the data from array 'q' column one

N = len(q)                                             # Establishes a value for 'N'
E_x = (1/N)*np.sum(Freq)                               # Sets up needed variables for the loop
E_y = (1/N)*np.sum(Volt)             
E_xx = (1/N)*np.sum(Freq**2)
y_1 = np.prod(q,axis=1)
E_xy = (1/N)*np.sum(y_1)    
m = ((E_xy)-(E_x)*(E_y))/((E_xx)-((E_x)**2))           # Equations for 'm' and 'c'
c = ((E_xx*E_y)-(E_x*E_xy))/(E_xx-(E_x**2))

SetRange = np.linspace(0.5e15,1.2e15,1000)             # Sets a range for the 'x' axis
CrvFit = ((m*(SetRange))+c)
plt.plot(SetRange,CrvFit,'r')                          # Plots the Best Fit line derived in the above loop

plt.ylabel('Voltage (V)')                              # Labels the axis
plt.xlabel('Frequency (Hz)')
plt.title('Millikan Results')                          # Gives the plot a title
plt.scatter(Freq,Volt)                                 # Displays the data as a scatterplot

# Part 5(e)

param0 = (0,100)
def LineFit(x,Slope,Y_Int):
    return (Slope*x+Y_Int)

FitParam, FitParamErr = spy.curve_fit(LineFit,Freq,Volt,param0)
FitY = LineFit(SetRange,FitParam[0],FitParam[1])
plt.plot(SetRange,FitY,'p')
plt.show()                                             # Skips the output data lines

```
