```python
import numpy as np
import matplotlib.pyplot as plt

a = np.loadtxt('sunspots.txt')                           # Loads in the data from the given file as an array

plt.scatter(a[:,0] , a[:,1] , 0.5)                       # Creates a scatterplot from the array and sets the point size
plt.xlabel('Occurences')                                 # Labels the axis
plt.ylabel('Index')
plt.title('Sunspot Data')                                # Gives the plot a title
plt.show()                                               # Displays the scatterplot
```
