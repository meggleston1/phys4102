HYDROGEN DATA ANALYSIS

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as spy

H_data1 = np.loadtxt("H_data1",float,skiprows=4,usecols=(1),delimiter=",",unpack=True)

def GausFit(x,xbar,sigma,A):
    return A*np.exp(-(x-xbar)**2/(2*sigma**2))

RngPkH1 = range(1130,1160)
WvLnPkH1 = 434

param0 = (1000,110,1)
FitParamH1,FitErrH1 = spy.curve_fit(GausFit,RngPkH1,H_data1[1130:1160],param0)

H_Fit1 = GausFit(RngPkH1,FitParamH1[0],FitParamH1[1],FitParamH1[2])

plt.plot(RngPkH1,H_Fit1)

print(FitParamH1)

plt.plot(RngPkH1,H_data1[1130:1160],'.')
plt.title("Hydrogen Data")
plt.ylabel("Voltage (Vdc)")
plt.xlabel("Time (Tenths of a Second)")
plt.xlim(1100,1200)
plt.ylim(0,1)
plt.show()
```

```python
RngPkH2 = range(2815,2850)
WvLnPkH2 = 486

param0 = (2830,40,5.5)
FitParamH2,FitErrH2 = spy.curve_fit(GausFit,RngPkH2,H_data1[2815:2850],param0)

H_Fit2 = GausFit(RngPkH2,FitParamH2[0],FitParamH2[1],FitParamH2[2])

plt.plot(RngPkH2,H_Fit2)

print(FitParamH2)

plt.plot(RngPkH2,H_data1[2815:2850],'.')
plt.title("Hydrogen Data")
plt.ylabel("Voltage (Vdc)")
plt.xlabel("Time (Tenths of a Second)")
plt.xlim(2800,2900)
plt.show()
```

```python
m = (WvLnPkH2 - WvLnPkH1)/(FitParamH2[0] - FitParamH1[0])
print(m)
b = WvLnPkH2 - m*FitParamH2[0]
print(b)

r = np.arange(0,len(H_data1))

def LinFit(m,b):
    return m*r + b

WavelengthH = LinFit(m,b)
plt.plot(WavelengthH,H_data1)
plt.title("Hydrogen Data")
plt.ylabel("Voltage (Vdc)")
plt.xlabel("Wavelength (nm)")
plt.show()
```
